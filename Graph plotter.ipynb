```python
import networkx as nx
import matplotlib.pyplot as plt
import re
```

```python
G = nx.Graph()
```

```python
file  = open('student/instances/small/3.lp', 'r') 
text = file.read() 
```

```python
vertexPattern = re.compile('vertex\\(\d+,\d+,\d+')
edgesPattern = re.compile('arc\\(\d+,\d+,\d+')
```

```python
for v in re.findall(vertexPattern, text):  
    number = re.findall('\d+',v)   
    G.add_node(number[0], pos=(int(number[1]), int(number[2])))   
```

```python
for e in re.findall(edgesPattern, text):  
    number = re.findall('\d+',e)   
    G.add_edge(number[0], number[1], r=number[2])  
```

```python
pos=nx.get_node_attributes(G,'pos')
```

```python
edge_labels = nx.get_edge_attributes(G,'r')
```

```python
plt.figure(1,figsize=(15,15)) 
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw_networkx_labels(G, pos)
nx.draw(G,pos)
plt.show()
```
